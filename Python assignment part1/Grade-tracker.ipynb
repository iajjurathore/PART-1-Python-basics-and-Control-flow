```python

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
#Step 1 — Clean#

cleaned_students = []

for student in raw_students:
    
    clean_name = student["name"].strip().title()
    # strip() removes leading/trailing spaces; title() fix case
    clean_roll = int(student["roll"])
    #convert to int
        
    marks_list = [int(m.strip()) for m in student["marks_str"].split(",")]
       # split marks string on ", " and cast each element to int
    
    cleaned = {"name": clean_name, "roll": clean_roll, "marks": marks_list}
    cleaned_students.append(cleaned)
    
#Step 2 — Validate#

  # Split name into words and check each word has only alphabetic chars

    words = clean_name.split()
    is_valid = all(word.isalpha() for word in words)

    if is_valid:
     print(f"✓ Valid name   -> {clean_name}")
    else:
     print(f"✗ Invalid name -> {clean_name}")

#Step 3 — Print Profile Card #  

    print("=" * 33)
    print(f"Student  : {clean_name}")
    print(f"Roll No  : {clean_roll}")
    print(f"Marks    : {marks_list}")
    print("=" * 33)
    print()


#Step 4 — Print Roll 103 in ALL CAPS & lowercase#

for s in cleaned_students:
    if s["roll"] == 103:
        print(f"ALL CAPS  : {s['name'].upper()}")
        print(f"lowercase : {s['name'].lower()}")
        break    
```

```python
#Task 2 — Marks Analysis Using Loops & Conditionals#

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
#step1 - For Loop:#
###################

for subject, mark in zip(subjects, marks):
    # determine grade based on mark
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
 
    # print subject, marks and grade on one line
    print(f"{subject:<10} -> Marks: {mark}  | Grade: {grade}")
```

```python
#Step 2 — Calculate Total, Average, Highest & Lowest#
####################################################

# total and average#
total   = sum(marks)
average = round(total / len(marks), 2)
 
# highest and lowest subject 
highest_mark    = max(marks)
highest_subject = subjects[marks.index(highest_mark)]
 
lowest_mark    = min(marks)
lowest_subject = subjects[marks.index(lowest_mark)]
 
print(f"Total Marks   : {total}")
print(f"Average Marks : {average}")
print(f"Highest       : {highest_subject} ({highest_mark})")
print(f"Lowest        : {lowest_subject} ({lowest_mark})")
```

```python
#Step 3 — While Loop:#
######################

new_count = 0
 
while True:
    new_subject = input("Enter subject name (or done to stop): ")
 
    # stop loop when user types done
    if new_subject.lower() == "done":
        break
 
    new_mark_str = input(f"Enter marks for {new_subject} (0-100): ")
 
    # validate: must be a number and between 0-100
    try:
        new_mark = int(new_mark_str)
        if 0 <= new_mark <= 100:
            subjects.append(new_subject)
            marks.append(new_mark)
            new_count += 1
        else:
            print("Warning: Marks must be between 0 and 100. Not added.")
    except ValueError:
        print("Warning: Invalid input. Please enter a number. Not added.")
 
# after loop: print summary
print(f"\nNew subjects added : {new_count}")
updated_avg = round(sum(marks) / len(marks), 2)
print(f"Updated average    : {updated_avg}")
```

#Task 3 — Class Performance Summary#
=

```python
###GIVEN DATA###

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
# Task 1: Compute average and result status
results = []
for name, scores in class_data:
    average = round(sum(scores) / len(scores), 2)
    status = "Pass" if average >= 60 else "Fail"
    results.append((name, average, status))

# Task 2: Print formatted class report
print(f"{'Name':<18} | {'Average':>7} | {'Status'}")
print("-" * 40)
for name, average, status in results:
    print(f"{name:<18} | {average:>7.2f} | {status}")

# Task 3: Print summary statistics
print()
passed = sum(1 for _, _, status in results if status == "Pass")
failed = len(results) - passed
print(f"Students Passed : {passed}")
print(f"Students Failed : {failed}")

topper = max(results, key=lambda x: x[1])
print(f"Class Topper    : {topper[0]} ({topper[1]})")

class_average = round(sum(avg for _, avg, _ in results) / len(results), 2)
print(f"Class Average   : {class_average}")
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# Step 1: Strip leading and trailing whitespace
clean_essay = essay.strip()
print( "Step 1 - Stripped Essay:",clean_essay)
print()

# Step 2: Convert to Title Case
title_essay = clean_essay.title()
print("Step 2 - Title Case:",title_essay)
print()

# Step 3: Count occurrences of "python" (case-insensitive)
count = clean_essay.count("python")
print("Step 3 - Count of 'python':",count)
print()

# Step 4: Replace "python" with "Python 🐍"
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Step 4 - Replaced Essay:",replaced_essay)
print()

# Step 5: Split into sentences by ". "
sentences = replaced_essay.split(". ")
print("Step 5 - Split into Sentences:",sentences)
print()

# Step 6: Print each sentence numbered, add "." if missing
print("Step 6 - Numbered Sentences:")
for i, sentence in enumerate(sentences, 1):
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{i}. {sentence}")
```
